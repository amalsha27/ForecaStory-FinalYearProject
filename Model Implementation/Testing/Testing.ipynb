```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Model names
models = ['LSTM', 'BiLSTM', 'CNN-LSTM', 'ARIMA-LSTM', 'BPNN-LSTM', 'ARIMA-CNNLSTM']
```

```python
# Metrics for each model
mae = [6.162847535, 3.360244704, 2.971011307, 8.889726135, 4.012468201, 4.513366411]
mape = [50.02269383, 31.82065284, 24.68082413, 75.63986517, 29.07295876, 26.40443499]
mse = [154.1552662, 67.16067883, 85.03576071, 413.3465594, 224.5709783, 863.6986972]
rmse = [12.41592792, 8.195161916, 9.22148365, 20.33092618, 14.98569245, 29.3887512]
```

```python
# Loss values for each model
loss = [408.0903625488281, 231.35470581054688, 618.928955078125, 193.4148712158203, 224.36793518066406, 1041.13916015625]
```

```python
x = np.arange(len(models))  # the label locations
width = 0.2  # the width of the bars
```

```python
fig, ax = plt.subplots()
rects1 = ax.bar(x - 3/2*width, mae, width, label='MAE')
rects2 = ax.bar(x - width/2, mape, width, label='MAPE')
rects3 = ax.bar(x + width/2, mse, width, label='MSE')
rects4 = ax.bar(x + 3/2*width, rmse, width, label='RMSE')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by model and metric')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation='vertical')  # Rotate model names
ax.legend()

fig.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(models, loss, color='purple')
plt.xlabel('Models')
plt.ylabel('Loss')
plt.title('Loss values for each model')
plt.xticks(rotation='vertical')
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(models, mae, marker='o', label='MAE')
plt.plot(models, mape, marker='o', label='MAPE')
plt.plot(models, mse, marker='o', label='MSE')
plt.plot(models, rmse, marker='o', label='RMSE')

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Scores by model')
plt.legend()
plt.xticks(rotation='vertical')

plt.show()
```
